## Cleaning using Imputation technique: Median

```python
# impute the missing values and create the missing value indicator variables for each numeric column.
df_numerical = Random_5_missing_DataTable.select_dtypes(include=np.number).copy()
numerical_cols = df_numerical.columns
print(numerical_cols)

for col in numerical_cols:
    df_numerical[col].fillna(df_numerical[col].median(), inplace=True)


df_non_numerical = Random_5_missing_DataTable.select_dtypes(exclude=np.number).copy()
non_numerical_cols = df_non_numerical.columns
print(non_numerical_cols)
for col in non_numerical_cols:
    df_non_numerical[col].fillna(df_non_numerical[col].mode()[0], inplace=True)
    

imputed_5_miss = pd.concat([df_non_numerical,df_numerical], axis=1, join='inner')
```

```python
# impute the missing values and create the missing value indicator variables for each numeric column.
df_numerical = Random_10_missing_DataTable.select_dtypes(include=np.number).copy()
numerical_cols = df_numerical.columns
print(numerical_cols)

for col in numerical_cols:
    df_numerical[col].fillna(df_numerical[col].median(), inplace=True)


df_non_numerical = Random_10_missing_DataTable.select_dtypes(exclude=np.number).copy()
non_numerical_cols = df_non_numerical.columns
print(non_numerical_cols)
for col in non_numerical_cols:
    df_non_numerical[col].fillna(df_non_numerical[col].mode()[0], inplace=True)
    

imputed_10_miss = pd.concat([df_non_numerical,df_numerical], axis=1, join='inner')
```

```python
# impute the missing values and create the missing value indicator variables for each numeric column.
df_numerical = Random_15_missing_DataTable.select_dtypes(include=np.number).copy()
numerical_cols = df_numerical.columns
print(numerical_cols)

for col in numerical_cols:
    df_numerical[col].fillna(df_numerical[col].median(), inplace=True)


df_non_numerical = Random_15_missing_DataTable.select_dtypes(exclude=np.number).copy()
non_numerical_cols = df_non_numerical.columns
print(non_numerical_cols)
for col in non_numerical_cols:
    df_non_numerical[col].fillna(df_non_numerical[col].mode()[0], inplace=True)
    

imputed_15_miss = pd.concat([df_non_numerical,df_numerical], axis=1, join='inner')
```

## Using SVM-lin Method for 5 % imputed dataset

```python
# Clean PeakTable and Extract PeakList
PercMiss = PeakTable['Perc_missing']  
PeakTableClean = PeakTable[(PercMiss < 20)] 
PeakList = PeakTableClean['Name']  

# Select Subset of Data
DataTable2 = imputed_5_miss[(imputed_5_miss.Class == 1) | (imputed_5_miss.Class == 0)]

# Create a Binary Y Vector 
Outcomes = DataTable2['Class']
Y = Outcomes.values 

# Split Data into Train (2/3) and Test (1/3)
DataTrain, DataTest, YTrain, YTest = train_test_split(DataTable2, Y, test_size=1/3, stratify=Y, random_state=85)

# Extract Train Data 
XTrain = DataTrain[PeakList]                                    
XTrainLog = np.log(XTrain)                                          
XTrainScale, mu, sigma = cb.utils.scale(XTrainLog, method='auto', return_mu_sigma=True)              
XTrainKnn = XTrainScale# cb.utils.knnimpute(XTrainScale, k=3)    

# Extract Test Data
XTest = DataTest[PeakList]                                     
XTestLog = np.log(XTest)                                          
XTestScale = cb.utils.scale(XTestLog, method='auto', mu=mu, sigma=sigma) 
XTestKnn = XTestScale #cb.utils.knnimpute(XTestScale, k=3)                                                                                                                 
```

## Optimization of Hyperparameter

```python
# Parameter Dictionary
C_range = [0.0001, 0.0005, 0.001, 0.005, 0.01, 0.05,  0.1]
param_dict = dict(C=C_range, kernel="linear")               

# Initialise
cv = cb.cross_val.KFold(model=cb.model.SVM,                      
                        X=XTrainKnn,                                 
                        Y=YTrain,                               
                        param_dict=param_dict,                   
                        folds=5,
                        n_mc=10)                            

# Run and Plot
cv.run()  
cv.plot(metric='auc')
cv.plot(metric='r2q2')
```

```python
# Parameter Dictionary
#C_range = [1e-6,1e-5,1e-4,5e-4,1e-3,5e-3,1e-2,1e-1]
C_range = [0.0001, 0.0002, 0.0003, 0.0004, 0.0005, 0.0006, 0.0007, 0.0008, 0.0009, 0.001]
param_dict = dict(C=C_range, kernel="linear")               

# Initialise
cv = cb.cross_val.KFold(model=cb.model.SVM,                      
                        X=XTrainKnn,                                 
                        Y=YTrain,                               
                        param_dict=param_dict,                   
                        folds=5,
                        n_mc=10)                                 

# Run and Plot
cv.run()  
cv.plot(metric='auc')
cv.plot(metric='r2q2')   
```

###  Build Model & Evaluate

```python
# Build Model
model = cb.model.SVM(C=0.0005, kernel="linear")     
YPredTrain = model.train(XTrainKnn, YTrain)
YPredTest = model.test(XTestKnn)

# Put YTrain and YPredTrain in a List
EvalTrain = [YTrain, YPredTrain]

# Put YTest and YPrestTest in a List
EvalTest = [YTest, YPredTest]

# Evaluate Model (include Test Dataset)
model.evaluate(testset=EvalTest) 
```

###  Bootstrap Evaluation

```python
# Extract X Data
XBoot = DataTable2[PeakList]
XBootLog = np.log(XBoot)
XBootScale = cb.utils.scale(XBootLog, method='auto')
XBootKnn = cb.utils.knnimpute(XBootScale, k=3)
YPredBoot = model.train(XBootKnn, Y)

# Build Boostrap Models
bootmodel = cb.bootstrap.Per(model, bootnum=100) 
bootmodel.run()

# Boostrap Evaluate Model (include Test Dataset)
bootmodel.evaluate(trainset=EvalTrain, testset=EvalTest)  
```

### Important features i.e. metabolites

```python
important_features = abs(model.model.coef_[0]) # extract coefficient
#get indices of those important features
number = important_features.argsort()
print(len(number))
number_new= number[::-1][:3]
print(number_new)
print()
#get more information from peaktable:
highest_metabolite = PeakList.iloc[number_new]
print("len of peak list=",len(PeakList))
print(highest_metabolite)
highest_metabolite_list = PeakTable.iloc[highest_metabolite.index-1]
#3 most important metabolites
highest_metabolite_list
```
